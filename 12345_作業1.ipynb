## 06. 集合
设由单词”paraparaparadise”和”paragraph”所生成的字符级bi-gram为分别为集合X
和集合Y
。求这两个集合的并(union)、交(intersection)、差(difference)。此外，请检查集合X
和Y
中是否包含bi-gram “se”。

```python
def bigram_self(text):
  return list(zip(*[text.split()[i:] for i in range(2)])),list(zip(*[text[i:] for i in range(2)]))

a = 'paraparaparadise'
b = 'paragraph'

X=set(bigram_self(a)[1])
Y=set(bigram_self(b)[1])

print(X | Y)
print(X & Y)
print(X - Y)
print(('s', 'e') in X)
print(('s', 'e') in Y)
```

## 20. 读取JSON文档
读取该包含Wikipedia文章的JSON文件，输出标题为”United Kingdom”的文章内容。基于本题的输出结果，完成第21至29问。

```python
import json
filename = 'enwiki-country.json'
with open(filename, mode='r') as f: #開檔案
  for line in f: #讀入
    line = json.loads(line)
    #print(line)
    if line['title'] == 'United Kingdom': # 找title 是 uk的那一行
      text_uk = line['text']
      break #結束
```

## 24. 提取引用的多媒体(jpg, PDF)
提取文章中所有被引用的多媒体文件。

```python
pattern = r'(File:(.*?)jpg|http:.+?pdf|File:(.*?)JPG)'
a=0
for i in re.findall(pattern, text_uk):
  a = a + 1
  if len(i[1])==0:
    print(a, i[0])
  else:
    print(a, i[1])
```

##25. 抽取Infobox信息
提取Infobox country中所有的字段名与其对应值，并将其存储为一个字典（dictionary）对象。

```python
# template
pattern = r'^\{\{Infobox country.*?$(.*?)^\}\}'
template = re.findall(pattern, text_uk, re.MULTILINE + re.DOTALL)

# template to dict
pattern = r'^\|(.+?)\s*=\s*(.+?)(?:(?=\n\|)|(?=\n$))'
result = dict(re.findall(pattern, template[0], re.MULTILINE + re.DOTALL)) #設dict把key:value聯合
for k, v in result.items():
  print(k + ': ' + v)
```
